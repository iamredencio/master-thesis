```python
# Step 1
# find all the basketball players that have an ID.
basketball_players = pd.read_csv('data/basketball_players.csv', ',')
basketball_players2 = pd.read_csv('data/basketball_players (2).csv', ',')

pb = []
pb += basketball_players.playerID.tolist()
pb += basketball_players2.playerID.tolist()
pb = list(set(pb))
```

```python
#Step 2, see how many players are in the files

count = 0
number_of_files = {}

all_files = [
basketball_awards_players,
basketball_hof,
basketball_master,
basketball_player_allstar,
basketball_players]

# Players,
# Seasons_Stats]

all_file_names = [
'basketball_awards_players',
'basketball_hof',
'basketball_master',
'basketball_player_allstar',
'basketball_players']

# 'Players',
# 'Seasons_Stats']

basketball_hof = basketball_hof.rename(columns={'hofID':'playerID'})
basketball_master = basketball_master.rename(columns={'bioID':'playerID'})
basketball_player_allstar = basketball_player_allstar.rename(columns={'player_id':'playerID'})
basketball_players = basketball_players.rename(columns={'player_id':'playerID'})
# Check how many players are in each file
for player in pb:
    for file, file_name in zip(all_files, all_file_names):  
        
        if player in list(set(file.playerID.tolist())):
            
            if file_name in number_of_files:
                number_of_files[file_name] += 1
            else:
                number_of_files[file_name] = 0
```

```python
#Step 3 get additional info of remaining players.
basketball_player_details = pd.merge(basketball_players, basketball_master.rename(columns={'bioID':'playerID'}), on='playerID')

# basketball_player_details.columns
```

```python
#Step 4 see which players has salaries
counter = 1

for p in list(set(basketball_players.playerID.tolist())):
    if p in list(set(salaries1978.player_id)):
        counter += 1

counter
```

```python
# Step 5 add salaries to dataset
player_and_salaries = pd.merge(basketball_player_details, salaries1978.rename(columns={'player_id':'playerID'}), on=['playerID', 'year'])
```

```python
# Step 6 Only select the basketball players with more than 5 years of playing
# The average career length of an NBA player is 4.8 years.

player_and_careerlength = player_and_salaries.groupby('playerID').filter(lambda x: len(x) >= 6).sort_values(['playerID', 'year'])
player_careerlenght_cols = player_and_careerlength.columns.tolist()
```

```python
player_and_careerlength.to_csv('results/player_and_careerlength.csv', index=False)
```
