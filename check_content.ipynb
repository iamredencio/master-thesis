```python
import math

all_files = [basketball_abbrev,
basketball_awards_coaches,
basketball_awards_players,
basketball_coaches,
basketball_draft,
basketball_hof,
basketball_master,
basketball_player_allstar,
basketball_players,
basketball_series_post,
basketball_teams,
nbaallelo,
ncaa_team_data,
Players,
Seasons_Stats]

all_file_names = ['basketball_abbrev',
'basketball_awards_coaches',
'basketball_awards_players',
'basketball_coaches',
'basketball_draft',
'basketball_hof',
'basketball_master',
'basketball_player_allstar',
'basketball_players',
'basketball_series_post',
'basketball_teams',
'nbaallelo',
'ncaa_team_data',
'Players',
'Seasons_Stats']

def check_content(file, name):
    player_data = {}
    alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

    for player, i in zip(file.playerID, range(len(file.playerID))):
        for col in file[file.playerID == player]:
            for item in file[file.playerID == player][col]:
                for team in file[file.playerID == player].tmID:
                    if item == item and not col == 'playerID': # is not NaN
                        item_str = ''

                        if isinstance(item, int):
                            item = str(item)
                            for let in item:
                                item_str += alphabet[int(let)]
                            item = item_str

                        player_data.setdefault('RowId #', []).append('Row ' + str(i))
                        player_data.setdefault('PlayerId', []).append(player) #Sentence
                        player_data.setdefault('Value', []).append(item) #Word
                        player_data.setdefault('Property', []).append(col) #POS
                        player_data.setdefault('Team', []).append(team)  #Tag
                        
    player_data = pd.DataFrame(player_data)
    player_data.to_csv('results/'+name+'.csv', ';', encoding='utf-8')       
    return pd.DataFrame(player_data)

for file, name in zip(all_files, all_file_names):
    player_data = check_content(file, name)
```
